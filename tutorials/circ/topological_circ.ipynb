```python
from topological_codes import TopologicalRegister, TopologicalCircuit, RepetitionQubit
from qiskit import execute, Aer, QuantumCircuit, QuantumRegister
```

```python
treg = TopologicalRegister(3) # Creates a Topological Register with 3 XXZZ qubits. 
circ = TopologicalCircuit(treg) # TopologicalCircuit uses the same circuit that created in TopologicalRegister 
                                # Allows us to use the same quantum circuit for all the different tqubits. 
circ.logical_x(treg[0])
# Could also use circ.logical_x(0) which also performs a logical_x on the Oth TopologicalQubit in the circuit we created 
circ.stabilize(treg[0])
circ.draw(output='mpl', fold=150)
```
